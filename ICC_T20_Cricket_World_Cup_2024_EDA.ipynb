### Teams win lost

```python
def get_team_results(df):
  all_teams = pd.unique(df[["Team 1", "Team 2"]].values.ravel("K"))

  team_results = {}

  for team in all_teams:
    team_results[team] = {'Matches': 0, 'Wins': 0, "Losses": 0, "No Result": 0, "Ties": 0}

  for index, row in df.iterrows():
    team1 = row["Team 1"]
    team2 = row["Team 2"]
    winner = row["Winner"]

    if winner == "no result":
      team_results[team1]["Matches"] += 1
      team_results[team2]["Matches"] += 1
      team_results[team1]["No Result"] += 1
      team_results[team2]["No Result"] += 1
    elif winner == "tied":
      team_results[team1]["Matches"] += 1
      team_results[team2]["Matches"] += 1
      team_results[team1]["Ties"] += 1
      team_results[team2]["Ties"] += 1
    elif winner == team1:
      team_results[team1]["Matches"] += 1
      team_results[team2]["Matches"] += 1
      team_results[team1]["Wins"] += 1
      team_results[team2]["Losses"] += 1
    elif winner == team2:
      team_results[team1]["Matches"] += 1
      team_results[team2]["Matches"] += 1
      team_results[team2]["Wins"] += 1
      team_results[team1]["Losses"] += 1


  return team_results


team_results_list = pd.DataFrame.from_dict(get_team_results(t20_wc_2024_match_results))
team_results_list = team_results_list.transpose().sort_values(by="Wins", ascending=False)
display(team_results_list.head(20))
```
